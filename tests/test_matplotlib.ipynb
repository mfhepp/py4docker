```python
# Import matplotlib and print the version
import matplotlib
import matplotlib.pyplot as plt
print("Matplotlib version:", matplotlib.__version__)

# Create a simple plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, marker='o')
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()

```
